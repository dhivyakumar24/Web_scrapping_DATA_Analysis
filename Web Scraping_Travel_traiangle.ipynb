# Web Scraping of TravelTriangle.com

```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
import warnings 
warnings.simplefilter('ignore', category=FutureWarning)
```

```python
https://traveltriangle.com/tour-packages/india/page/2
```

```python
packages=[]
package_names = []
Days =[]
discount=[]
Price= []
Actual_Price = []
cities= []
Cities_names=[]
hotel= []
pattern="(\d+)% Off"
pattern2="â\x82¹\s?([\d,]+)"
pattern3= "â\x82¹\s?([\d,]+)"
pattern4= "([A-Za-z\s]+)"
for i in range(1,40):
    url = f"https://traveltriangle.com/tour-packages/india/page/{i}"
    print (url)
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    packages=[]
    for i in soup.find_all("span",class_="fw9 m0 f16 pfc3"):
       packages.append(i.text)
    for package in packages:
        cleaned_name = re.sub(r"(?:From .+|For .+|: .+|Go on .+|Discover .+)", "", package).strip()
        if cleaned_name == "":
            package_names.append(np.nan)  # Pandas NA for missing values
        else:
            package_names.append(cleaned_name)
    for i in soup.find_all("span",class_="iblock sfc6"):
        Days.append(i.text)

    for i in soup.find_all("span",class_="f12 fw4 pt2 pb2 pl8 pr8 radius20 pbc1 sfcw at_discount_label"):
        regex=re.findall(pattern,i.text)
        if len(regex) >0:
            discount.append(regex[0])
        else:
            discount.append(np.nan)

    for i in soup.find_all("h4",class_="sfc3 m0 f24 fw9 priceVal at_newprice"):
        regex=re.findall(pattern2,i.text)
        if len(regex) >0:
            price = int(regex[0].replace(",", ""))
            Price.append(price)
        else:
            Price.append(np.nan)
    for i in soup.find_all("h4", itemprop="priceSpecification"):
        regex = re.findall(pattern3, i.text)  # Find all matched prices
        if len(regex) > 1:  # Ensure there are at least two prices in the list
            second_price = regex[1]  # Get the second price
            second_price_clean = int(second_price.replace(",", ""))  # Remove commas and convert to integer
            Actual_Price.append(second_price_clean)
        else:
            Actual_Price.append(np.nan)

    for i in soup.find_all("ul",class_="clearfix package-cities-list at_packagecity_list"):
        cities.append(i.text)
        Cities_names = [line.split(' ')[0] for line in cities]

    for i in soup.find_all("div", class_="flex alignCenter"):
        text = i.text.strip()  # Get the text and remove leading/trailing whitespace
        if text:  # If text is present, append it
            hotel.append(text)
        else:  # If text is empty, append NaN
            hotel.append(np.nan)

```

```python
df = pd.DataFrame({"Package Name": package_names,"No.OF_Days":Days,"Actual_price":Actual_Price,"Package_Price":Price,"City":Cities_names,"Other_City":cities})
df["Discount_amt"] = df["Actual_price"] - df["Package_Price"]

# Print the DataFrame
df
```

```python
df.to_excel("cities.xlsx", index=False)
```

```python
l=[]
links = []
for i in range(1,40):
    url = f"https://traveltriangle.com/tour-packages/india/page/{i}"
    print (url)
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    for i in soup.find_all('a'):
        if i.has_attr('href'):
            if len(re.findall("\/packages.*\d$",i['href']))>0:
                l.append(re.findall("\/packages.*\d$",i['href'])[0])
    for i in l:
        links.append("https://traveltriangle.com"+i)
```

```python
rate= []
total_traveled=[]
for i in newlink:
    page1=requests.get(i)
    soup1=BeautifulSoup(page1.text)
    for i in soup1.find_all("p",class_="f12 at_package_count"):
        rate.append(i.text)
    for i in soup1.find_all("p",class_="pfc4 f12 text-center m8"):
        total_traveled.append(i.text)
```

```python
Rate=[]
for item in rate:
    match = re.search(r"Rated (\d\.\d)", item)
    if match:
        Rate.append (match.group(1))
        #print(f"Rated: {rated_score}")
    else:  # If text is empty, append NaN
        Rate.append(np.nan)
```
